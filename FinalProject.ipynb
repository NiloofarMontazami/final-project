## Data Cleaning

Here I extracted **month**,**season**,**weather description**,**weekends** and **wind speed** out of the columns and added them as a new column in the dataframe,additionally I removed some columns that were not really needed for my model.

```python
# Import the pandas library
import pandas as pd
import datetime



# Define a function to extract the month from the date
def extract_month(date_str):
    # Parse the date string using the strptime() method
    date = datetime.datetime.strptime(date_str, "%Y-%m-%d")
    
    # Extract the month from the date
    month = date.strftime("%m")
    
    # Return the month as an integer value
    return int(month)

# Apply the extract_month() function to the 'date' column
month = df["date"].apply(extract_month)

# Add the extracted month as a new column in the data frame using the assign() method
df = df.assign(month=month)



# Create a list with the names of the columns to be dropped
columns_to_drop = ["avgtempF", "maxtempF", "mintempF", "uvIndex","astronomy"]

# Use the drop() method to remove the specified columns from the data frame
df= df.drop(columns_to_drop, axis=1)

import datetime

# Define a function to extract the season from the date
def extract_season(date_str):
    # Parse the date string using the strptime() method
    date = datetime.datetime.strptime(date_str, "%Y-%m-%d")
    
    # Extract the month from the date
    month = date.strftime("%m")
    
    # Identify the season based on the month
    if month in ["12", "01", "02"]:
        season = "Winter"
    elif month in ["03", "04", "05"]:
        season = "Spring"
    elif month in ["06", "07", "08"]:
        season = "Summer"
    else:
        season = "Autumn"
    
    # Return the season
    return season

# Apply the extract_season() function to the 'date' column
season = df["date"].apply(extract_season)

# Add the extracted season as a new column in the data frame using the assign() method
df= df.assign(season=season)


# Define a function to extract the season from the date
def is_weekend(date_str):
    # Parse the date string using the strptime() method
    date = datetime.datetime.strptime(date_str, "%Y-%m-%d")
    
    # Extract the day of the week from the date
    day = date.strftime("%A")
    
    # Check if the date is a weekend (Saturday or Sunday)
    if day in ["Saturday", "Sunday"]:
        return True
    else:
        return False

# Apply the is_weekend() function to the 'date' column
weekend = df["date"].apply(is_weekend)

# Add the extracted season as a new column in the data frame using the assign() method
df= df.assign(weekend=weekend)

# Define a function to extract the weather description from the hourly data
def get_weather_desc(hourly_data):
    # Get the first dictionary from the list of hourly data
    hour = hourly_data[0]
    
    # Extract the weather description from the dictionary
    weather_desc = hour["weatherDesc"]
    
    # Return the weather description as a string
    return weather_desc

# Apply the get_weather_desc() function to the 'hourly' column
weather_desc = df["hourly"].apply(get_weather_desc)

# Add the extracted weather description as a new column in the data frame using the assign() method
df= df.assign(weather_desc=weather_desc)

# Define a function to extract the windspeed in kilometers per hour from the hourly data
def get_windspeed_kmph(hourly_data):
    # Get the first dictionary from the list of hourly data
    hour = hourly_data[0]
    
    # Extract the windspeed in kilometers per hour from the dictionary
    windspeed_kmph = hour["windspeedKmph"]
    
    # Return the windspeed in kilometers per hour as a string
    return windspeed_kmph

# Apply the get_windspeed_kmph() function to the 'hourly' column
windspeed_kmph = df["hourly"].apply(get_windspeed_kmph)

# Add the extracted windspeed in kilometers per hour as a new column in the data frame using the assign() method
df= df.assign(windspeed_kmph=windspeed_kmph)

df
```

I created a new column to count the number of delayed flights

```python
# Group the data by 'Date' and count the number of rows in each group
delayed_count = df_merged.groupby('Date').size()

# Convert the result to a dataframe
delayed_count = pd.DataFrame({'Delayed_Count': delayed_count})

# Merge the 'delayed_count' dataframe with the original dataframe
df = df_merged.merge(delayed_count, left_on='Date', right_index=True)

df
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot of delayed_count and windspeed
plt.figure(figsize=(10, 8))
plt.scatter(df['windspeed_kmph'], df['Delayed_Count'])

# Add labels to the axes
plt.xlabel('Windspeed')
plt.ylabel('Number of Delayed Flights')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group the data by month
groups = df.groupby('month')

# Create a figure with multiple subplots
fig, axs = plt.subplots(nrows=4, ncols=3, sharex=True, sharey=True, figsize=(30, 30))

# Loop through each group and create a scatter plot
for ax, (month, group) in zip(axs.flat, groups):
    ax.scatter(group['windspeed_kmph'], group['Delayed_Count'])
    ax.set_title(month)

# Show the plot
plt.show()

```
